# Part 1 - Basic statistics of the data

```python
#Importing lbraries

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#Read CSV File

df=pd.read_csv('6153237444115dat.csv')
df
```

```python
# read the data with na_values assigned
df = pd.read_csv('6153237444115dat.csv', na_values=['*', '**', '***', '****', '*****', '******'])

# replace "not available" with NaN
df.replace('not available', pd.NA, inplace=True)
df
```

```python
#create function to check for missing values in the data

def count_of_null(df):
    count=df.isnull().sum().sum()
    return count

# print missing value count in the data
count_null = count_of_null(df)
print(count_null)
```

```python
#create function to check duplicate values in the data
def check_duplicates(df):
    count=df.duplicated().sum().sum()
    return count

# print duplicate values in the data
count_duplicates = check_duplicates(df)
print(count_duplicates)
```

```python
#create a function to check information about the data
def about_data (df):
        about=df.info()
        return about
    
#print information about the data
info=about_data (df)
print(info)
```

```python
# create function to print unique values in the dataframe

def print_unique_values(df):
    for column in df.columns:
        unique_values = df[column].unique()
        print(f"Column '{column}' has {len(unique_values)} unique values:")
        print(unique_values)
print_unique_values(df)
```

```python
#Create function to check statistical summary of the dataset
def stat_summary (df):
    statistics_sum = df.describe()
    return statistics_sum
#print statistical summary of data
summary = stat_summary (df)
summary
```

```python
# calculate the mean Fahrenheit temperature in the 'TEMP' column
mean_temp = df['TEMP'].mean()

print(f'The mean Fahrenheit temperature in the data is: {mean_temp:.2f}')
```

```python
# calculate the standard deviation of the Maximum temperature in the 'MAX' column
std_max = df['MAX'].std()

print(f'The standard deviation of the Maximum temperature in the data is: {std_max:.2f}')
```

```python
#assign num_stations as number of unique stations
#print number of unique stations
num_stations = df['USAF'].unique()

print(f'The number of unique stations in the data are two stations: {num_stations}')
```

# Part 2 - Data manipulation

```python
#create a new dataframe called selected that only contains the columns specified: USAF, YR--MODAHRMN, TEMP, MAX, and MIN
selected = df[['USAF', 'YR--MODAHRMN', 'TEMP', 'MAX', 'MIN']]
selected
```

```python
#drop all rows that have a missing value in the TEMP column and save the updated dataframe back into selected.
selected = selected.dropna(subset=['TEMP'])
selected
```

```python
#print the unique values in MIN column
selected.MIN.unique()
```

```python
#drop all rows that have a missing value in the MIN column and save the updated dataframe back into selected.
selected = selected.dropna(subset=['MIN'])
selected
```

```python
#drop all rows that have a missing value in the MAX column and save the updated dataframe back into selected.
selected = selected.dropna(subset=['MAX'])
selected
```

```python
#print unique values in max column
selected.MAX.unique()
```

```python
#apply this formula to the 'TEMP' column and create a new column called 'Celsius'.
#ignore warnings
import warnings
from pandas.core.common import SettingWithCopyWarning
warnings.simplefilter(action="ignore", category=SettingWithCopyWarning)

selected.loc[:, 'Celsius'] = (selected['TEMP'] - 32) / 1.8
selected
#create a new column called 'Celsius' in the 'selected' DataFrame, 
#which contains the Celsius temperature values converted from Fahrenheit
```

```python
#use the round() method of pandas DataFrame to round the values in the 'Celsius' column to 1 decimal place:

selected['Celsius'] = selected['Celsius'].round(1)
selected
```

# Part 3 - Data selection 

```python
#use boolean indexing to select all rows where the USAF code is 29980 
#and assign them to a new variable called kumpula as follows:

kumpula = selected[selected['USAF'] == 29980]
kumpula
```

```python
#use boolean indexing to select the rows where the USAF code is 28450 as follows:
rovaniemi = selected[selected['USAF'] == 28450]
rovaniemi
```

```python
#save the kumpula DataFrame to a CSV file
kumpula.to_csv('Kumpula_temps_May_Aug_2017.csv', index=False, sep=',', float_format='%.1f')
```

```python
#save the rovaniemi Dataframe to a CSV file
rovaniemi.to_csv('Rovaniemi_temps_May_Aug_2017.csv', sep=',', float_format='%.1f', index=False)
```

# Part 4 - Data analysis

```python
#median temperature in Celsius for Helsinki Kumpula, use the median() function on the "Celsius" column of the kumpula df:
kumpula_median_temp = kumpula['Celsius'].median()
print("The median temperature in Celsius for Helsinki Kumpula was:", round(kumpula_median_temp, 1), "°C")
```

```python
# median temperature in Celsius in Rovaniemi, we can use the median method on the Celsius column of the rovaniemi df:
rovaniemi_median = rovaniemi['Celsius'].median()
print(f"The median temperature in Celsius in Rovaniemi was {rovaniemi_median:.1f}°C.")
```

```python
# Calculate mean, min, max for Kumpula in May
kumpula_may_stats = kumpula_may.describe()[['TEMP']].loc[['mean', 'min', 'max']]
print("Kumpula Mean, Min, Max temperatures for May:\n", kumpula_may_stats)

# Calculate mean, min, max for Kumpula in June
kumpula_june_stats = kumpula_june.describe()[['TEMP']].loc[['mean', 'min', 'max']]
print("Kumpula Mean, Min, Max temperatures for June:\n", kumpula_june_stats)
```

```python
# Mean, min, and max temperatures for Rovaniemi in May
rovaniemi_may_mean = rovaniemi_may['Celsius'].mean()
rovaniemi_may_min = rovaniemi_may['Celsius'].min()
rovaniemi_may_max = rovaniemi_may['Celsius'].max()

# Mean, min, and max temperatures for Rovaniemi in June
rovaniemi_june_mean = rovaniemi_june['Celsius'].mean()
rovaniemi_june_min = rovaniemi_june['Celsius'].min()
rovaniemi_june_max = rovaniemi_june['Celsius'].max()

# Print the results
print("Rovaniemi temperatures in May:")
print("Mean temperature: {:.1f}°C".format(rovaniemi_may_mean))
print("Minimum temperature: {:.1f}°C".format(rovaniemi_may_min))
print("Maximum temperature: {:.1f}°C".format(rovaniemi_may_max))
print("\n")
print("Rovaniemi temperatures in June:")
print("Mean temperature: {:.1f}°C".format(rovaniemi_june_mean))
print("Minimum temperature: {:.1f}°C".format(rovaniemi_june_min))
print("Maximum temperature: {:.1f}°C".format(rovaniemi_june_max))
```
